```python
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import math 
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(42)
kangaroo_population = np.random.normal(loc=10, scale=3, size=10000)

num_samples = 50
iterations = 1000

running_means = []
for i in range(1, iterations + 1):
    sample = np.random.choice(kangaroo_population, size=num_samples, replace=True)
    running_means.append(np.mean(sample))

plt.figure(figsize = (10,5))
plt.plot(range(1, iterations + 1), running_means, label="Sample Average", color="blue")
plt.axhline(np.mean(kangaroo_population), color='red', linestyle='--', label="True Population Mean")
plt.title(f"Law of Large Numbers Demonstration\nSample Size = {num_samples}, Iterations = {iterations}")
plt.xlabel("Number of Iterations")
plt.ylabel("Average Age of Kangaroos")
plt.legend()
plt.show()
```
